```python
# 딕셔너리 자료형 Key, Value 구성
dic = {'name': 'pey', 'phone': '010-9999-1234', 'birth': '1118'}
print(dic)
print(type(dic))
print(dic['name']) # Key값을 부르면 Value값이 리턴 됌
```

```python
dic['name'] = 'sam'
print(dic['name'])

dic['sex'] = 'man' # key값 추가
print(dic)

del dic['phone']  # key값 지우기
print(dic)
```

```python
print(dic.keys())   # Key값 출력
print(dic.values()) # Value값 출력
```

```python
a = {1:'a', 1:'b'}  # 같은 key값을 주면 덮어쓴다
print(a)
```

```python
a = [1,2,3]
print (1 in a)
print (4 in a)

b = {1:'a', 2:'b'}
print( 'a' in b)          # Key값에 a가 있는지 확인
print( 'a' in b.values()) # Value에 a가 있는지 확인
```

```python
# 집합 자료형
# 인덱스가 없어서 인덱싱이 불가능! 인덱싱으로 접근하려면 리스트나 튜플로 변환해야함
# 중복도 불가능!
s1 = set([1,2,3,3])
print(s1)
print(type(s1))

l1 =list(s1)
print(l1[0])

l2 = tuple(s1)
print(l2[1])
```

```python
s1 = set([1, 2, 3, 4, 5, 6])
s2 = set([4, 5, 6, 7, 8, 9])

l1 = set(s1)
l2 = set(s2)
l3 = l1 & l2    # 교집합
print (l3)

print( s1 | s2) # 합집합
print( s1 - s2) # 차집합
```

```python
l1 = [1,2,1,2,3,3,5,5,5,5]
l2 = list(set(l1))    # set으로 중복 제거
print (l2)
```
