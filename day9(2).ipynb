```python
from PyQt5.QtCore import Qt
from PyQt5.QtWidgets import (QWidget, QApplication, QLabel, QLineEdit, QTextEdit, QGridLayout, QListWidget,
                             QVBoxLayout, QPushButton, QHBoxLayout)
from PyQt5.QtGui import QFont, QColor, QPalette
import sys
import requests

# Weather 정보 가져오는 함수
class Weather:
    def __init__(self, dt_txt, temp, feels_like, temp_min, temp_max, humidity, main):
        self.dt_txt = dt_txt
        self.temp = temp
        self.feels_like = feels_like
        self.temp_min = temp_min
        self.temp_max = temp_max
        self.humidity = humidity
        self.main = main

    def __str__(self):
        return f"{self.dt_txt} - {self.main} - {self.temp}°C (Feels like: {self.feels_like}°C)"

def call_weather():
    API_key = "9e56e0954e2b4912bd3b297fcd3fae47"
    city_name = "Paju"
    url_weather = f"https://api.openweathermap.org/data/2.5/forecast?q={city_name}&appid={API_key}"
    url_weather += "&units=metric"
    response_w = requests.get(url_weather)
    result_w = response_w.json()
    
    weather_list = []
    for i in range(40):
        root = result_w["list"]
        temp = root[i]["main"]["temp"]
        feels_like = root[i]["main"]["feels_like"]
        temp_min = root[i]["main"]["temp_min"]
        temp_max = root[i]["main"]["temp_max"]
        humidity = root[i]["main"]["humidity"]
        dt_txt = root[i]["dt_txt"]
        main = root[i]["weather"][0]["main"]
        
        weather = Weather(dt_txt, temp, feels_like, temp_min, temp_max, humidity, main)
        weather_list.append(weather)
    return weather_list

# PyQt5 UI 디자인
class MyApp(QWidget):
    def __init__(self):
        super().__init__()
        self.initUI()

    def initUI(self):
        # Window title and size
        self.setWindowTitle("날씨 정보 🌤️")
        self.setGeometry(600, 200, 700, 600)
        
        # Set the background color of the widget
        palette = self.palette()
        palette.setColor(QPalette.Background, QColor(240, 240, 240))  # Light gray background
        self.setAutoFillBackground(True)
        self.setPalette(palette)

        # Set the font for all labels and buttons
        font = QFont("Arial", 12)
        
        # Main Layout (Vertical box layout)
        vbox = QVBoxLayout()
        vbox.setSpacing(10)  # Adjust spacing between elements

        # Title label with emoji
        title_label = QLabel("날씨 정보 🌤️", self)
        title_label.setFont(QFont("Arial", 18, QFont.Bold))
        title_label.setAlignment(Qt.AlignCenter)
        title_label.setStyleSheet("color: #2c3e50; margin-bottom: 10px;")

        # Button for requesting weather info with emoji
        button_call = QPushButton("날씨 요청 ☀️", self)
        button_call.setFont(QFont("Arial", 14))
        button_call.setStyleSheet("background-color: #3498db; color: white; padding: 10px; border-radius: 5px;")
        
        # List widget to display the weather data
        qlist = QListWidget(self)
        qlist.setFont(QFont("Arial", 12))
        qlist.setStyleSheet("""
            QListWidget {
                background-color: white;
                border: 1px solid #ddd;
                border-radius: 5px;
                padding: 10px;
            }
            QListWidget::item {
                padding: 10px;
                margin-bottom: 5px;
                border-bottom: 1px solid #f0f0f0;
            }
            QListWidget::item:hover {
                background-color: #ecf0f1;
            }
        """)

        # Add widgets to the layout
        vbox.addWidget(title_label)
        vbox.addWidget(button_call)
        vbox.addWidget(qlist)

        self.setLayout(vbox)

        # Connect button click to the weather request
        def button_clicked():
            qlist.clear()
            weather_list = call_weather()
            for weather in weather_list:
                # Add weather icon to the list item
                weather_icon = self.get_weather_icon(weather.main)
                qlist.addItem(f"{weather_icon} {weather}")

        button_call.clicked.connect(button_clicked)
        
        # Show window
        self.show()

    # 날씨에 맞는 이모티콘 반환 함수
    def get_weather_icon(self, weather_main):
        if weather_main == "Clear":
            return "☀️"
        elif weather_main == "Clouds":
            return "☁️"
        elif weather_main == "Rain":
            return "🌧️"
        elif weather_main == "Snow":
            return "❄️"
        elif weather_main == "Thunderstorm":
            return "⚡"
        elif weather_main == "Drizzle":
            return "🌦️"
        else:
            return "🌈"  # Default

if __name__ == "__main__":
    app = QApplication(sys.argv)
    window = MyApp()
    sys.exit(app.exec_())


```
