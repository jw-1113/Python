```python
marks = [90, 25, 67, 45, 80]
for number in range(len(marks)):
    if marks[number] < 60:
        continue
    print("%d번 학생 축하합니다. 합격입니다." % (number+1))
```

```python
a = [1,2,3,4]
b = []
for i in a:
  b.append(i * 3)
print(b)

result = [i * 3 for i in a] # list compregension
print(result)

result2 = [i * 3 for i in a if i%2 == 0]
print(result2)
```

```python
# zip함수 두개 이상의 리스트를 동시에 순회
names = ['홍길동', '김철수', '이영희']
korean = [85, 92, 78]
english = [90, 88, 95]

for name, kor, eng in zip(names, korean, english):
     print(f"{name}: 국어 {kor}점, 영어 {eng}점")
```

```python
def sub(a,b):
  return a - b

print(sub(10, 5))
print(sub(a=20,b=5))  # 매개변수를 지정해서 값을 줄때는 순서 상관x
print(sub(b=5,a=20))
```

```python
def say_myself(name, age, man=True):  # man = diffult 매개변수, 맨뒤에만 사용가능
    print("나의 이름은 %s 입니다." % name)
    print("나이는 %d살입니다." % age)
    if man:
        print("남자입니다.")
    else:
        print("여자입니다.")

say_myself("노재웅",27)
say_myself("아이유",30,False)
```

```python
def add_many(*args):
    result = 0
    for i in args:
         result = result + i   # *args에 입력받은 모든 값을 더한다.
    return result

print(add_many(1,2,3,4,5))
```

```python
def print_kwargs(**kwargs): # 입력받은 값을 딕셔너리 형태로 출력
  print(kwargs)

print_kwargs(name='foo', age=3)
```

```python
def say_nick(nick):
  if nick == "바보":
    return
  print(f"나의 별명은 {nick}입니다.")

print(say_nick("바보"))
```

```python
a = 1               # 전역변수
def vartest(a):     # 지역변수
    return a + 1

print(vartest(a))
print(a)
```

```python
# global 명령어로 진입할수 있지만 사용하지 않음
a = 1
def vartest():
    global a
    a = a+1

vartest()
print(a)
```

```python
# 리스트가 변경가능한(mutable) 자료형이기 때문에 값이 변경된다
def change_list(my_list):
  my_list.append(4)

a = [1,2,3]
change_list(a)
print(a)
```

```python
def two_times(x):
  return x*2

print(list(map(two_times, [1,2,3,4,])))
print(list(map(lambda x : x*2, [1,2,3,4,])))  # lamda함수로 간편하게 사용가능
```
