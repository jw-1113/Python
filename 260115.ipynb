```python
a= [1,4,3,2]
a.sort()
print(a)

a.sort(reverse=True)
print(a)

a.reverse()
print(a)
```

```python
a = [1,4,3,2]
print(a.index(3))
print(a.index(2))

a.insert(1,3)
print(a)
a.remove(3) # 가까이에 있는 인덱스의 3을 지움
print(a)

keep = a.pop() # 맨마지막 요소 뽑기
print(a)
print(keep)

keep = a.pop(2) # 2번인덱스 요소 뽑기
print(a)
print(keep)
```
