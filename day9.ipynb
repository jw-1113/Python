```python
# 계산기 프로그램 
from PyQt5.QtCore import Qt
from PyQt5.QtWidgets import ( QWidget, QApplication, QLabel, QLineEdit,
                                QTextEdit, QGridLayout, QPushButton )
import sys

class MyApp(QWidget):
    def __init__(self):
        super().__init__()
        self.initUI()

    def initUI(self):
        # 초기화면 구성
        grid = QGridLayout()
        self.result = ""

        value_list = ["7", "8", "9", "*",
                     "4", "5", "6", "-",
                     "1", "2", "3", "+",
                     "C", "0", "=", "/"]
        
        button_list = []
        for i in range(16):
            button = QPushButton()
            button.clicked.connect(self.button_cliked)
            button_list.append(button)

        for i, button in enumerate(button_list):
            button.setText(value_list[i])

        self.ql_result = QLabel("0", alignment=Qt.AlignRight)
        grid.addWidget(self.ql_result, 0, 0, 1, 4)
        # addWidget (행, 열, 차지할 행, 차지할 열)
        count = 0
        for i in range(1, 5):
            for j in range(4):
                grid.addWidget(button_list[count], i, j)
                count += 1

        self.setLayout(grid)
        self.setGeometry(300, 300, 300, 200)

        self.setStyleSheet("""
            /* 전체 윈도우 스타일 */
            QWidget {
                background-color: #2E2E2E; /* 어두운 배경색 */
            }

            /* 결과창 (QLabel) 스타일 */
            QLabel {
                font-size: 48px;
                color: #FFFFFF; /* 흰색 텍스트 */
                background-color: #2E2E2E;
                border: none;
                padding: 10px;
            }

            /* 모든 버튼 (QPushButton)의 기본 스타일 */
            QPushButton {
                font-size: 20px;
                font-weight: bold;
                color: #FFFFFF;
                background-color: #505050; /* 중간 회색 */
                border: none;
                border-radius: 30px; /* 둥근 모서리 */
                min-height: 60px; /* 최소 높이 */
                min-width: 60px; /* 최소 너비 */
            }
            QPushButton:hover {
                background-color: #6a6a6a; /* 마우스 올렸을 때 */
            }
            QPushButton:pressed {
                background-color: #7f7f7f; /* 클릭했을 때 */
            }

            /* 연산자 버튼(*, -, +, /) 스타일 */
            QPushButton#operator {
                background-color: #FF9500; /* 주황색 */
            }
            QPushButton#operator:hover {
                background-color: #ffab3c;
            }
            QPushButton#operator:pressed {
                background-color: #ffc476;
            }

            /* C (Clear) 버튼 스타일 */
            QPushButton#clear {
                background-color: #D4D4D2; /* 밝은 회색 */
                color: #000000; /* 검은색 텍스트 */
            }
            QPushButton#clear:hover {
                background-color: #e2e2e1;
            }
            QPushButton#clear:pressed {
                background-color: #f1f1f1;
            }
        """)

        self.show()

    def button_cliked(self):
        # __init__, initUI()와 같은 level
        # ~~~~~~~~
        # 버튼 3을 눌렀다 -> QLabel의 text가 3이 되게
        # button_cliked 나를 부른 사람이 누구야?
        button = self.sender() # 버튼 3이 누르면 3버튼
        key = button.text()

        if key == "=":
            cal_result = eval(self.result)
            self.ql_result.setText(str(cal_result))
            self.result = ""
        elif key == "C":
            self.result = ""
            self.ql_result.setText("0")
        else:
            self.result += key
            self.ql_result.setText(self.result)
        
if __name__ == "__main__": # 내가 직접 실행한다면
    app = QApplication.instance()
    if app is None:
        app = QApplication(sys.argv)
    exam = MyApp()
    sys.exit(app.exec_())
```

```python
# 1 to 9 게임 만들기
from PyQt5.QtCore import Qt
from PyQt5.QtWidgets import ( QWidget, QApplication, QLabel, 
                            QLineEdit, QTextEdit, QGridLayout,
                            QVBoxLayout, QPushButton)
from PyQt5.QtGui import QFont
import sys
import time
import random

class MyApp(QWidget):
    def __init__(self):
        super().__init__()
        self.n = 4
        self.n2 = pow(self.n, 2)
        self.initUI()

    def initUI(self):
        self.answer = 1 # 내가 눌러야 할 숫자
        self.setWindowTitle(f"1 to {self.n2} GAME")
        self.setGeometry(300, 300, 300, 400)
        
        vbox = QVBoxLayout()
        self.ql_result = QLabel(f"1 to {self.n2}", alignment=Qt.AlignCenter)
        self.ql_result.setFont(QFont("", 24))
        
        self.button_start = QPushButton("GAME START")
        self.button_start.setFixedHeight(50)
        self.button_start.clicked.connect(self.start_game)

        self.grid = QGridLayout()

        self.button_list = []
        for i in range(self.n2):
            button = QPushButton()
            button.setMinimumSize(90, 60)
            button.setStyleSheet("font-size:20px;")
            button.setEnabled(False) # 처음에 못 누르게!!
            self.button_list.append(button)

        count = 0
        for i in range(self.n):
            for j in range(self.n):
                self.grid.addWidget(self.button_list[count], i, j)
                count += 1

        vbox.addWidget(self.ql_result)
        vbox.addWidget(self.button_start)
        
        vbox.addLayout(self.grid)

        self.setLayout(vbox)
        self.show()

    def start_game(self):
        self.button_start.setEnabled(False)

        # 1 ~ 9까지 랜덤하게 들어 있는 자료
        number_list = list(range(1, self.n2 + 1))
        import random
        random.shuffle(number_list)
        
        for i in range(self.n2):
            self.button_list[i].setEnabled(True)
            self.button_list[i].setText(f"{str(number_list[i])}")
            self.button_list[i].clicked.connect(self.button_cliked)
        self.start_time = time.perf_counter() # 시작 시각 입력

    def button_cliked(self):
        button = self.sender()
        if int(button.text()) == self.answer:
            button.setEnabled(False)
            button.setText("")
            self.answer += 1

            if self.answer == self.n2 + 1:
                self.end_time = time.perf_counter() # 종료 시각 입력

                record_time = self.end_time - self.start_time
                self.ql_result.setText(f"{record_time:.2f}sec")

                self.button_start.setEnabled(True)
                self.answer = 1

if __name__ == "__main__": # 내가 직접 실행한다면
    app = QApplication.instance()
    if app is None:
        app = QApplication(sys.argv)
    exam = MyApp()
    sys.exit(app.exec_())
```

```python
from PyQt5.QtCore import Qt
from PyQt5.QtWidgets import QWidget, QApplication, QLabel, QGridLayout, QVBoxLayout, QPushButton
from PyQt5.QtGui import QFont
import sys
import time
import random

class MyApp(QWidget):
    def __init__(self):
        super().__init__()
        self.n = 6
        self.n2 = pow(self.n, 2)
        self.initUI()

    def initUI(self):
        self.answer = 1  # 내가 눌러야 할 숫자
        self.setWindowTitle(f"1 to {self.n2} GAME")
        self.setGeometry(300, 300, 450, 550)

        # 기본 레이아웃 설정
        vbox = QVBoxLayout()

        # 타이틀 레이블
        self.ql_result = QLabel(f"1 to {self.n2} GAME", alignment=Qt.AlignCenter)
        self.ql_result.setFont(QFont("Arial", 28, QFont.Bold))
        self.ql_result.setStyleSheet("color: #2d3436;")

        # 타이머 레이블
        self.timer_label = QLabel("00.00 sec", alignment=Qt.AlignCenter)
        self.timer_label.setFont(QFont("Arial", 18))
        self.timer_label.setStyleSheet("color: #e74c3c; margin-top: 10px;")

        # 게임 시작 버튼
        self.button_start = QPushButton("GAME START")
        self.button_start.setFixedHeight(70)
        self.button_start.setStyleSheet("""
            background-color: #f39c12;
            color: white;
            font-size: 22px;
            border-radius: 12px;
            padding: 10px 20px;
            border: none;
        """)
        self.button_start.clicked.connect(self.start_game)

        # 버튼들 그리드 설정
        self.grid = QGridLayout()
        self.button_list = []

        for i in range(self.n2):
            button = QPushButton()
            button.setMinimumSize(90, 90)
            button.setStyleSheet("""
                font-size: 20px;
                background-color: #ecf0f1;
                border-radius: 12px;
                border: none;
                color: #34495e;
                transition: background-color 0.3s ease, transform 0.2s ease;
            """)
            button.setEnabled(False)  # 게임 시작 전엔 버튼 비활성화
            self.button_list.append(button)

        # 그리드 레이아웃에 버튼들 배치
        count = 0
        for i in range(self.n):
            for j in range(self.n):
                self.grid.addWidget(self.button_list[count], i, j)
                count += 1

        # 레이아웃에 추가
        vbox.addWidget(self.ql_result)
        vbox.addWidget(self.timer_label)
        vbox.addWidget(self.button_start)
        vbox.addLayout(self.grid)

        self.setLayout(vbox)
        self.setStyleSheet("background-color: #f5f5f5;")  # 배경색 설정
        self.show()

    def start_game(self):
        self.button_start.setEnabled(False)
        self.timer_label.setText("00.00 sec")  # 타이머 리셋

        # 1부터 n^2까지의 랜덤 숫자들
        number_list = list(range(1, self.n2 + 1))
        random.shuffle(number_list)

        # 각 버튼에 숫자 배치 및 클릭 이벤트 연결
        for i in range(self.n2):
            self.button_list[i].setEnabled(True)
            self.button_list[i].setText(f"{str(number_list[i])}")
            self.button_list[i].clicked.connect(self.button_cliked)

        self.start_time = time.perf_counter()  # 게임 시작 시간 기록

    def button_cliked(self):
        button = self.sender()

        # 올바른 버튼 클릭 시 처리
        if int(button.text()) == self.answer:
            button.setEnabled(False)
            button.setText("")

            # 클릭 시 버튼 색상과 스타일 변경
            button.setStyleSheet("""
                background-color: #2ecc71;  # 초록색으로 변경
                color: white;
                font-size: 22px;
                border-radius: 12px;
                transform: scale(1.1);  # 클릭된 버튼이 살짝 커짐
            """)

            # 클릭된 버튼이 커졌다가 원래 크기로 돌아오는 효과
            button.setStyleSheet("""
                background-color: #2ecc71;
                color: white;
                font-size: 22px;
                border-radius: 12px;
                transform: scale(1.1);
                transition: transform 0.1s ease-out;
            """)

            self.answer += 1

            # 게임 종료 조건 (모든 숫자를 다 눌렀을 때)
            if self.answer == self.n2 + 1:
                self.end_time = time.perf_counter()  # 종료 시간 기록

                record_time = self.end_time - self.start_time
                self.timer_label.setText(f"{record_time:.2f} sec")

                self.button_start.setEnabled(True)
                self.answer = 1

if __name__ == "__main__": 
    app = QApplication.instance()
    if app is None:
        app = QApplication(sys.argv)
    exam = MyApp()
    sys.exit(app.exec_())
```
