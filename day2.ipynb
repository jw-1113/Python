```python
# 반복문
# while : 대개 반복 횟수가 정해져 있지 않을 때
# for   : 대개 반복 횟수가 정해져 있을때

# while
# while 조건식:
#     실행문장

total = 0
# num = int(input("자연수를 입력하세요: "))
# total += num
# num = int(input("자연수를 입력하세요: "))
# total += num
# num = int(input("자연수를 입력하세요: "))
# total += num
# num = int(input("자연수를 입력하세요: "))
# total += num
# print(f"누적 합: {total}")

num = int(input("자연수를 입력하세요: "))
total = 0
while num > 0:
    total += num
    print(f"누적 합: {total}")

    num = int(input("자연수를 입력하세요: "))
```

```python
# 사용자가 자연수를 입력
# 입력받은 자연수를 누적합 후 출력
# 단, 음수를 입력하면 입력 종료!

# break 키워드
# 가장 가까운 반복문을 탈출한다!

# 조건식 자리에 True
# 무한 반복문 : 탈출 조건 명시 break 사용
total = 0

while True:
    num = int(input("자연수를 입력하세요: "))

    if num < 0:
        break
    
    total += num
    print(f"누적 합: {total}")
    
```

```python
current_weight = int(input("현재 몸무게 : "))
goal_weight = int(input("목표 몸무게 : "))

week = 1

while True:
    loss_weight = int(input(f"{week}주차 감량 몸무게 : "))
    current_weight -= loss_weight

    if current_weight <= goal_weight:
        break

    week += 1
    
print(f"{current_weight}kg 달성!! 축하합니다!")
```
